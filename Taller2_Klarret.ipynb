# ***Serie de Fourier (Exponencial, Trigonom√©trica y Compacta)***

***Se√±ales:*** Se aplica a se√±ales peri√≥dicas en tiempo continuo.
***Espectro:*** Discreto. Representa la se√±al como una suma de componentes sinusoidales (o exponenciales complejas) en frecuencias discretas que son m√∫ltiplos de la frecuencia fundamental de la se√±al.
***Semejanzas:*** Todas las formas de la serie de Fourier representan la misma se√±al peri√≥dica. La forma exponencial es la m√°s fundamental y las formas trigonom√©trica y compacta pueden derivarse de ella.
***Diferencias:*** La forma trigonom√©trica usa senos y cosenos, la exponencial usa exponenciales complejas y la compacta usa amplitudes y fases.
Transformada de Fourier

***Se√±ales:*** Se aplica a se√±ales no peri√≥dicas en tiempo continuo.
***Espectro:*** Continuo. Representa la se√±al en todas las frecuencias posibles.
***Semejanzas:*** Es una extensi√≥n de la serie de Fourier para se√±ales no peri√≥dicas. En el l√≠mite, cuando el per√≠odo de una se√±al peri√≥dica tiende a infinito, la serie de Fourier se convierte en la transformada de Fourier.
***Diferencias:*** Trabaja con se√±ales no peri√≥dicas y produce un espectro continuo, a diferencia de la serie de Fourier que trabaja con se√±ales peri√≥dicas y produce un espectro discreto.

# ***Transformada de Fourier en Tiempo Discreto (DTFT)***

***Se√±ales:*** Se aplica a se√±ales no peri√≥dicas en tiempo discreto.
***Espectro:*** Continuo y peri√≥dico en la frecuencia. El espectro se repite cada 2œÄ2œÄ radianes.
***Semejanzas:*** Es el an√°logo de la transformada de Fourier para se√±ales de tiempo discreto.
***Diferencias:*** Trabaja con se√±ales de tiempo discreto y su espectro es continuo y peri√≥dico, a diferencia de la transformada de Fourier que trabaja con se√±ales de tiempo continuo y su espectro es continuo y no peri√≥dico.

# ***Transformada Discreta de Fourier (DFT)***

***Se√±ales:*** Se aplica a se√±ales peri√≥dicas en tiempo discreto (o a una secuencia finita de puntos que se asumen peri√≥dicos).
***Espectro:*** Discreto. Calcula el espectro en un n√∫mero finito de frecuencias discretas.
***Semejanzas:*** Es la versi√≥n discreta de la transformada de Fourier. Puede verse como una aproximaci√≥n de la DTFT calculada en un n√∫mero finito de puntos. Es fundamental para el procesamiento digital de se√±ales.
***Diferencias:*** Trabaja con se√±ales de tiempo discreto (finitas) y produce un espectro discreto, a diferencia de la DTFT que trabaja con se√±ales de tiempo discreto (infinitas o finitas) y produce un espectro continuo y peri√≥dico.

# ***Algoritmo Fast Fourier Transform (FFT)***

***Consiste en:*** La FFT es un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT) y su inversa. No es una nueva transformada, sino una forma r√°pida de calcular la DFT.
Utilidad: Su principal utilidad es reducir dr√°sticamente el costo computacional del c√°lculo de la DFT. Esto ha sido fundamental para el desarrollo de numerosas aplicaciones en procesamiento digital de se√±ales, como compresi√≥n de audio e im√°genes, an√°lisis de espectro, filtrado digital, etc.
***Explicaci√≥n del Algoritmo:*** El algoritmo FFT, en particular la versi√≥n de Cooley-Tukey (la m√°s com√∫n), se basa en el principio de "divide y vencer√°s". Descompone la DFT de tama√±o NN en DFTs m√°s peque√±as, t√≠picamente de tama√±o N/2N/2. Esto se logra explotando las propiedades de periodicidad y simetr√≠a de las ra√≠ces de la unidad complejas utilizadas en la definici√≥n de la DFT. El proceso de descomposici√≥n se repite recursivamente hasta llegar a DFTs de tama√±o 1, que son triviales de calcular. Luego, los resultados de estas peque√±as DFTs se combinan para obtener el resultado final. Hay diferentes implementaciones de la FFT, como la de radix-2 (para tama√±os NN que son potencias de 2), o radix-4, entre otras.

***Costo Computacional:***

DFT directa: Calcular la DFT directamente utilizando la definici√≥n matem√°tica requiere aproximadamente O(N2) operaciones complejas (multiplicaciones y sumas), donde N es el n√∫mero de puntos.
FFT: Para un tama√±o N que es una potencia de 2, el algoritmo FFT de radix-2 reduce el costo computacional a aproximadamente O(Nlog2N) operaciones complejas.
Comparaci√≥n: La diferencia en el costo computacional es muy significativa, especialmente para valores grandes de N. Por ejemplo, para N=1024, la DFT directa requiere aproximadamente (1024)2‚âà1 mill√≥n de operaciones, mientras que la FFT requiere aproximadamente 1024√ólog2(1024)=1024√ó10=10240 operaciones. Esto hace que la FFT sea miles de veces m√°s r√°pida que el c√°lculo directo de la DFT para tama√±os t√≠picos utilizados en el procesamiento de se√±ales.
