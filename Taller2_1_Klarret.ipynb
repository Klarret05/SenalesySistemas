# ***Serie de Fourier (Exponencial, Trigonom√©trica y Compacta)***

***Se√±ales:*** Se aplica a se√±ales peri√≥dicas en tiempo continuo.
***Espectro:*** Discreto. Representa la se√±al como una suma de componentes sinusoidales (o exponenciales complejas) en frecuencias discretas que son m√∫ltiplos de la frecuencia fundamental de la se√±al.
***Semejanzas:*** Todas las formas de la serie de Fourier representan la misma se√±al peri√≥dica. La forma exponencial es la m√°s fundamental y las formas trigonom√©trica y compacta pueden derivarse de ella.
***Diferencias:*** La forma trigonom√©trica usa senos y cosenos, la exponencial usa exponenciales complejas y la compacta usa amplitudes y fases.
Transformada de Fourier

***Se√±ales:*** Se aplica a se√±ales no peri√≥dicas en tiempo continuo.
***Espectro:*** Continuo. Representa la se√±al en todas las frecuencias posibles.
***Semejanzas:*** Es una extensi√≥n de la serie de Fourier para se√±ales no peri√≥dicas. En el l√≠mite, cuando el per√≠odo de una se√±al peri√≥dica tiende a infinito, la serie de Fourier se convierte en la transformada de Fourier.
***Diferencias:*** Trabaja con se√±ales no peri√≥dicas y produce un espectro continuo, a diferencia de la serie de Fourier que trabaja con se√±ales peri√≥dicas y produce un espectro discreto.

# ***Transformada de Fourier en Tiempo Discreto (DTFT)***

***Se√±ales:*** Se aplica a se√±ales no peri√≥dicas en tiempo discreto.
***Espectro:*** Continuo y peri√≥dico en la frecuencia. El espectro se repite cada 2œÄ2œÄ radianes.
***Semejanzas:*** Es el an√°logo de la transformada de Fourier para se√±ales de tiempo discreto.
***Diferencias:*** Trabaja con se√±ales de tiempo discreto y su espectro es continuo y peri√≥dico, a diferencia de la transformada de Fourier que trabaja con se√±ales de tiempo continuo y su espectro es continuo y no peri√≥dico.

# ***Transformada Discreta de Fourier (DFT)***

***Se√±ales:*** Se aplica a se√±ales peri√≥dicas en tiempo discreto (o a una secuencia finita de puntos que se asumen peri√≥dicos).
***Espectro:*** Discreto. Calcula el espectro en un n√∫mero finito de frecuencias discretas.
***Semejanzas:*** Es la versi√≥n discreta de la transformada de Fourier. Puede verse como una aproximaci√≥n de la DTFT calculada en un n√∫mero finito de puntos. Es fundamental para el procesamiento digital de se√±ales.
***Diferencias:*** Trabaja con se√±ales de tiempo discreto (finitas) y produce un espectro discreto, a diferencia de la DTFT que trabaja con se√±ales de tiempo discreto (infinitas o finitas) y produce un espectro continuo y peri√≥dico.

# ***Algoritmo Fast Fourier Transform (FFT)***

***Consiste en:*** La FFT es un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT) y su inversa. No es una nueva transformada, sino una forma r√°pida de calcular la DFT.
Utilidad: Su principal utilidad es reducir dr√°sticamente el costo computacional del c√°lculo de la DFT. Esto ha sido fundamental para el desarrollo de numerosas aplicaciones en procesamiento digital de se√±ales, como compresi√≥n de audio e im√°genes, an√°lisis de espectro, filtrado digital, etc.
***Explicaci√≥n del Algoritmo:*** El algoritmo FFT, en particular la versi√≥n de Cooley-Tukey (la m√°s com√∫n), se basa en el principio de "divide y vencer√°s". Descompone la DFT de tama√±o NN en DFTs m√°s peque√±as, t√≠picamente de tama√±o N/2N/2. Esto se logra explotando las propiedades de periodicidad y simetr√≠a de las ra√≠ces de la unidad complejas utilizadas en la definici√≥n de la DFT. El proceso de descomposici√≥n se repite recursivamente hasta llegar a DFTs de tama√±o 1, que son triviales de calcular. Luego, los resultados de estas peque√±as DFTs se combinan para obtener el resultado final. Hay diferentes implementaciones de la FFT, como la de radix-2 (para tama√±os NN que son potencias de 2), o radix-4, entre otras.

***Costo Computacional:***

DFT directa: Calcular la DFT directamente utilizando la definici√≥n matem√°tica requiere aproximadamente O(N2) operaciones complejas (multiplicaciones y sumas), donde N es el n√∫mero de puntos.
FFT: Para un tama√±o N que es una potencia de 2, el algoritmo FFT de radix-2 reduce el costo computacional a aproximadamente O(Nlog2N) operaciones complejas.
Comparaci√≥n: La diferencia en el costo computacional es muy significativa, especialmente para valores grandes de N. Por ejemplo, para N=1024, la DFT directa requiere aproximadamente (1024)2‚âà1 mill√≥n de operaciones, mientras que la FFT requiere aproximadamente 1024√ólog2(1024)=1024√ó10=10240 operaciones. Esto hace que la FFT sea miles de veces m√°s r√°pida que el c√°lculo directo de la DFT para tama√±os t√≠picos utilizados en el procesamiento de se√±ales.

### Modulaci√≥n de Amplitud con Detecci√≥n Coherente

La modulaci√≥n de amplitud (AM) es un m√©todo para enviar informaci√≥n, como el sonido, mont√°ndola sobre una onda de radio de mayor frecuencia que act√∫a como "portadora". Cuando usamos la detecci√≥n coherente (tambi√©n llamada s√≠ncrona), el receptor necesita tener una copia id√©ntica de esa onda portadora original, tanto en su velocidad de oscilaci√≥n (frecuencia) como en su punto de inicio (fase). Esto es crucial para poder "descodificar" la se√±al y recuperar la informaci√≥n que se envi√≥.

El truco est√° en que el receptor multiplica la se√±al AM que le llega por su propia versi√≥n de la onda portadora. Despu√©s de esta multiplicaci√≥n, se usa un "filtro pasabajos" para eliminar las frecuencias altas indeseadas, dejando solo la se√±al de audio o datos original, ¬°lista para ser escuchada o procesada!

La gran ventaja de este m√©todo es su impresionante capacidad para ignorar el ruido y otras interferencias. Como el receptor est√° perfectamente sincronizado con la se√±al original, puede distinguir muy bien lo que queremos de lo que no. Sin embargo, no todo es perfecto: su principal inconveniente es que es bastante complicado y caro de implementar, ya que conseguir esa sincronizaci√≥n perfecta de la onda portadora en el receptor no es tarea f√°cil.

Matem√°ticamente, una se√±al AM com√∫n (de doble banda lateral con portadora) se representa as√≠:

$$s(t) = A_c [1 + k_a m(t)] \cos(2\pi f_c t)$$

Donde:
* $A_c$ es la amplitud de la portadora.
* $m(t)$ es la se√±al mensaje.
* $f_c$ es la frecuencia de la portadora.
* $k_a$ es la sensibilidad de amplitud del modulador.
* El **√≠ndice de modulaci√≥n** ($m$) es $m = k_a \cdot A_m$, donde $A_m$ es la amplitud m√°xima de la se√±al mensaje.

---

### Aplicaciones Principales

Dado que la detecci√≥n coherente es compleja, se reserva para situaciones donde la calidad del sonido y la resistencia al ruido son absolutamente cr√≠ticas:

* Radio AM Est√©reo: La escuchamos en sistemas como el
C-QUAM, que permite separar los canales de audio izquierdo y derecho para un sonido est√©reo en la radio AM.
* Comunicaciones de Datos: En algunos tipos de m√≥dems lentos y sistemas de transmisi√≥n de datos donde cada bit de informaci√≥n es vital.
* Sistemas de Radar y Telemetr√≠a: Es fundamental en aplicaciones donde se necesita extraer informaci√≥n muy precisa de se√±ales que pueden ser d√©biles o estar muy afectadas por el ruido.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

def plot_signals(t, msg, carrier, am_signal, modulation_index, msg_type):
    """
    Funci√≥n para graficar las se√±ales en tiempo y frecuencia.
    """
    # --- Dominio del Tiempo ---
    plt.figure(figsize=(12, 8))
    plt.suptitle(f'Modulaci√≥n AM con Se√±al {msg_type} (√çndice de Modulaci√≥n: {modulation_index})', fontsize=16)

    plt.subplot(3, 1, 1)
    plt.title('Se√±al Mensaje')
    plt.plot(t, msg, 'b')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.title('Se√±al Portadora')
    plt.plot(t, carrier, 'g')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.subplot(3, 1, 3)
    plt.title('Se√±al Modulada AM')
    plt.plot(t, am_signal, 'r')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    # --- Dominio de la Frecuencia ---
    N = len(t)
    T = t[1] - t[0]

    yf_msg = rfft(msg)
    xf_msg = rfftfreq(N, T)

    yf_am = rfft(am_signal)
    xf_am = rfftfreq(N, T)

    plt.figure(figsize=(12, 6))
    plt.suptitle(f'Espectro de Frecuencia (FFT) - Se√±al {msg_type}', fontsize=16)

    plt.subplot(2, 1, 1)
    plt.title('Espectro de la Se√±al Mensaje')
    plt.plot(xf_msg, np.abs(yf_msg), 'b')
    plt.ylabel('|M(f)|')
    plt.xlim(0, fc * 0.1) # Zoom en las frecuencias bajas para la se√±al mensaje
    plt.grid(True)

    plt.subplot(2, 1, 2)
    plt.title('Espectro de la Se√±al Modulada AM')
    plt.plot(xf_am, np.abs(yf_am), 'r')
    plt.xlabel('Frecuencia (Hz)')
    plt.ylabel('|S(f)|')
    plt.xlim(0, fc * 2) # Visualizar hasta el doble de la frecuencia portadora
    plt.grid(True)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()


# --- Par√°metros Generales ---
sampling_rate = 10000  # Tasa de muestreo (Hz)
t = np.arange(0, 1.0, 1/sampling_rate) # Vector de tiempo de 1 segundo

# Frecuencias
fc = 500  # Frecuencia de la portadora (Hz)
fm = 10   # Frecuencia de la se√±al mensaje (para coseno)

# Amplitudes
Ac = 1.0  # Amplitud de la portadora
Am = 1.0  # Amplitud de la se√±al mensaje

# --- Elige el √≠ndice de modulaci√≥n ---
# m < 1: Submodulaci√≥n
# m = 1: Modulaci√≥n cr√≠tica
# m > 1: Soblemodulaci√≥n
modulation_index = float(input("Introduce el √≠ndice de modulaci√≥n (ej. 0.5, 1.0, 1.5): "))

# Sensibilidad del modulador (ka) se ajusta para obtener el √≠ndice deseado
ka = modulation_index / Am

# --- Se√±al Portadora ---
carrier = Ac * np.cos(2 * np.pi * fc * t)

# ===============================================
#  EJEMPLO 1: SE√ëAL MENSAJE TIPO PULSO RECTANGULAR
# ===============================================
msg_pulse = np.zeros_like(t)
# Crear un pulso rectangular entre t=0.2s y t=0.4s
msg_pulse[(t > 0.2) & (t < 0.4)] = Am
am_signal_pulse = Ac * (1 + ka * msg_pulse) * np.cos(2 * np.pi * fc * t)

plot_signals(t, msg_pulse, carrier, am_signal_pulse, modulation_index, "Pulso Rectangular")

# ===============================================
#  EJEMPLO 2: SE√ëAL MENSAJE TIPO COSENO
# ===============================================
msg_cosine = Am * np.cos(2 * np.pi * fm * t)
am_signal_cosine = Ac * (1 + ka * msg_cosine) * np.cos(2 * np.pi * fc * t)

plot_signals(t, msg_cosine, carrier, am_signal_cosine, modulation_index, "Cosenoidal")
```
